```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year-Month'] = df['Date'].dt.to_period('M')
monthly = (df.groupby('Year-Month')
.agg({'Quantity': 'sum',
'Revenue' : 'sum',
'Cost'     : 'sum'})
.round(2))
monthly.index = monthly.index.to_timestamp()   # makes it a real date index
monthly = monthly.sort_index()
```
